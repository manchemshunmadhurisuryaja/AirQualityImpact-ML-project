```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
```

```python
tree=['Logistic Regression','Naive Bayes','SVM','Optimizble NN']
precision = (87.62,89.35,85.92,91.73)
recall = (75.44,77.55,81.32,92.20)
f1 = (79.37,81.32,86.19,91.92)
accuracy = (92.20,92.20,96.60,98.00)
xpos=np.arange(len(precision))
plt.style.use('default')
plt.figure(figsize=(7,6))
barwidth=0.2
plt.bar(xpos,precision,color='cornflowerblue',width=barwidth,label='precision')
plt.bar(xpos+0.2,recall,color='salmon',width=barwidth,label='recall')
plt.bar(xpos+0.4,f1,color='lightgreen',width=barwidth,label='f1')
plt.bar(xpos+0.6,accuracy,color='plum',width=barwidth,label='accuracy')
plt.xticks(xpos + 0.2, tree)
plt.xlabel('ML techniques')
plt.ylabel('Percentage(%)')
plt.title('Performance Metrics for Comparison of various ML Techniques')
plt.ylim(70, 100)
plt.legend(frameon=False, loc='upper center', ncol=4, bbox_to_anchor=(0.5, -0.15), bbox_transform=plt.gca().transAxes)
```
